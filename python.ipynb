### STEP 2
1. Gathering data from our users
***

```python
x1=[]
y1=[]
name=str(input("Please enter a title for this graph:"))
xname=str(input("Please enter the name of the x axis:"))
yname=str(input("Please enter the name of the y axis:"))
n=int(input("Please enter how many points you'd like to plot:"))
for i in range(0,n):
    xu=int(input("Please Enter the x value:"))
    yu=int(input("Please Enter the y value:"))
    x1.append(xu)
    y1.append(yu)
yname=yname+"-->"
xname=xname+"-->"
print("Data entry done!")
```

### STEP 3
1. Importing all the modules for the app
2. Drawing the graph 
3. Displaying the graph
***

```python
import matplotlib.pyplot as plt
import numpy as np 
plt.xlabel(xname,fontweight="bold")
plt.ylabel(yname,fontweight="bold")
plt.title(name,fontweight="bold")
plt.plot(x1,y1,linestyle='dotted',color="red",marker='o',markersize=8)
plt.grid()
plt.show()
```

```python
import numpy as np 
fig,ax=plt.subplots(2)
ax[0].spines["bottom"].set_position("center")
ax[0].spines["left"].set_position("center")
ax[0].spines["right"].set_color('None')
ax[0].spines["top"].set_color("None")
ax[0].xaxis.set_ticks_position('bottom')
ax[0].yaxis.set_ticks_position('left')
x=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.arange(-2*np.pi,2*np.pi,0.01)
ax[0].plot(x,np.sin(x),"-",color="red")
ax[1].plot(y,np.cos(x),"--",color="blue")
```
